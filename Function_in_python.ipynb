Vamos a usar las siguiente librerias

```python
import matplotlib.pyplot as plt #libreria para graficar
import numpy as np              #libreria para manejo de vectores y utilidades matematicas
```

Funciones Algebraicas

# 1) FUNCION LINEAL

```python
def f(m,x,b):
  return m*x + b

res = 100

m = 1

b = 5

x = np.linspace(-10.0, 10.0, num=res)

y =f(m,x,b)

fig, ax = plt.subplots()

ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color = 'r')
ax.axvline(x=0, color = 'r')

```

# 2) Funciones Polinomicas 
      2.1)parabola
      2.2)polinomio
      

```python
def f(x):
  return x**2

res = 100

x = np.linspace(-10.0, 10.0, num=res)  #desde -10 hasta 10 con 100 numeros de muestra

y =f(x)                                #la funcion 

fig, ax = plt.subplots()              #

ax.plot(x,y)
ax.grid()
ax.axhline(y=0, color = 'r')
ax.axvline(x=0, color = 'r')
```

```python
def f(x):
  return 2*x**8 -x**4 +3*x**2 + 4

res = 100

x = np.linspace(-10.0, 10.0, num=res)  #desde -10 hasta 10 con 100 numeros de muestra

y =f(x)                                #la funcion 

fig, ax = plt.subplots()              #definimos esto 

ax.plot(x,y)                          #hacemos la grafica
ax.grid()                             #creamos rendijas
ax.axhline(y=0, color = 'r')          # hacemos lienas a los ejes 
ax.axvline(x=0, color = 'r')

plt.xlim([-10,10])          #para hacer un set de los ejes y controlar la grfica
```

# 3) Funciones transcendentes
      3.1)sen cos and tang
      3.2)Funcion exponencial and logaritmo
      3.3) Funciones seccionadas

```python
def f(x):
  return np.cos(x)

res = 100                             #La resolucion o sea cuantos puntos voy a usar para graficar entre mas mejor

x = np.linspace(-10.0, 10.0, num=res)  #desde -10 hasta 10 con 100 numeros o puntos de muestra

y =f(x)                                #la funcion 

fig, ax = plt.subplots()              #definimos esto 

ax.plot(x,y)                          #hacemos la grafica
ax.grid()                             #creamos rendijas
ax.axhline(y=0, color = 'r')          # hacemos lienas a los ejes 
ax.axvline(x=0, color = 'r')

plt.xlim([-2,10])          #para hacer un set de los ejes y controlar la grfica
```

3.2) Funcion exponencial

```python
def f(x):
  return np.e**x

res = 100

x = np.linspace(-1.0, 1.0, num=res)  #desde -10 hasta 10 con 100 numeros de muestra

y =f(x)                                #la funcion 

fig, ax = plt.subplots()              #definimos esto 

ax.plot(x,y)                          #hacemos la grafica
ax.grid()                             #creamos rendijas
ax.axhline(y=0, color = 'r')          # hacemos lienas a los ejes 
ax.axvline(x=0, color = 'r')

plt.xlim([-1,1])          #para hacer un set de los ejes y controlar la grfica
```

3.2) Funcion logaritmo

```python
def f(x):
  return np.log2(x)

res = 100

x = np.linspace(0.01, 256, num=res)  #desde -10 hasta 10 con 100 numeros de muestra

y =f(x)                                #la funcion 

fig, ax = plt.subplots()              #definimos esto 

ax.plot(x,y)                          #hacemos la grafica
ax.grid()                             #creamos rendijas
ax.axhline(y=0, color = 'r')          # hacemos lienas a los ejes 
ax.axvline(x=0, color = 'r')

plt.xlim([0,250])          #para hacer un set de los ejes y controlar la grfica
```

3.3) Funcion seccionada

```python
def H(x):                          #funcion que recibe un x
  y = np.zeros(len(x))             #creamos un vector lleno de ceros
  for idx, x in enumerate(x):      #hacemos un bucle para poner posicion y valor
    if x>=0:                       #la condicion de la funcion a trozos
      y[idx]= 1.0                   
  return y

res = 100

x = np.linspace(-10.0, 10.0, num=res)  #desde -10 hasta 10 con 100 numeros de muestra

y = H(x)                                #la funcion 

fig, ax = plt.subplots()              #definimos esto 

ax.plot(x,y)                          #hacemos la grafica
ax.grid()                             #creamos rendijas


plt.xlim([-10,10])          #para hacer un set de los ejes y controlar la grfica
```
