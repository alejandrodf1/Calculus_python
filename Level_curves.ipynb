# Level curves

```
from matplotlib import cm #to manage colors
import numpy as np
import matplotlib.pyplot as plt
```

```
def z(x,y):
    return np.sin(x) + 2*np.cos(y)

res = 100      #resolution or numbers of points to use
 
x = np.linspace(-4,4,num = res)     #we create the values for x axes
y = np.linspace(-4,4,num = res)     #we create the values for y axes

x, y = np.meshgrid(x,y)             #generate the pair of points

z = z(x, y)                         #we call the function

fig, ax = plt.subplots(subplot_kw={'projection': '3d'})   #we create the graph with a 3d surface

suf =ax.plot_surface(x,y,z, cmap = cm.cool)               #we pass the values for the graph
fig.colorbar(suf)                                         #to create a gradient bar
```

Now let's create the level curve:

```
fig2, ax2 = plt.subplots()              #we create the 2d surface that is default
level_map = np.linspace(np.min(z),np.max(z), num =100 )   #we create the curves, we can change num to 

cp = ax2.contour(x,y,z, level_map, cmpa= cm.cool)        #now we draw the figures contour
#cp = ax2.contourf(x,y,z, level_map, cmpa= cm.cool)      #contourf is another way to graph this level curves

fig2.colorbar(cp)                                        #we create the gradient 
```
